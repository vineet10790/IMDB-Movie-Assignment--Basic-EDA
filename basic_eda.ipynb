- ### Subtask 3.5: Find the best directors

    1. Group the dataframe using the `director_name` column.
    2. Find out the top 10 directors for whom the mean of `imdb_score` is the highest and store them in a new dataframe `top10director`.  Incase of a tie in IMDb score between two directors, sort them alphabetically. 

```python
# Write your code for extracting the top 10 directors here

top10director=movies.groupby('director_name')['imdb_score'].mean().sort_values(ascending=False).head(10)

top10director=pd.DataFrame(top10director)   # extracting top 10 directors

# writing code to create a dataframe which will be sorted by mean imdb score and 
# Incase of a tie in IMDb score between two directors, sort them alphabetically.

top10director.reset_index(inplace=True)      
top10director["imdb_score"]=top10director["imdb_score"].apply(lambda x: round(x,2))
top10director_final=pd.DataFrame()                        # Initializing an empty dataframe
list1=list(set(list(top10director["imdb_score"])))        # creating a list of mean imdb scores having unique values
list1.sort(reverse=True)
for item in list1:
    df=top10director.loc[top10director["imdb_score"]==item,["director_name","imdb_score"]]
    df.sort_values(['director_name'],inplace=True)        # appending alphabetically sorted dataframes to top10director_final dataframe
    top10director_final=top10director_final.append(df)

top10director_final=top10director_final.rename(columns={"imdb_score":"mean_imdb_score"})

top10director_final
```

-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
import numpy as np
import pandas as pd
movies=pd.read_csv("C:/Users/pc/Desktop/assignment_29_04_2020/Movie+Assignment+Data.csv")

list_genre_1=[]     #declaring two empty lists which will receive the first 2 genres from each item in column(genres)
list_genre_2=[]

for i in range(len(movies.index)):
    
    sample=movies.loc[i,"genres"].split("|") #creating a temporay list sample which will receive the split item in column genres
    
    if len(sample)>1:
        
        list_genre_1.append(sample[0])    #row in column genres which have more than 2 genres, the first 2 can be allocated to each lists
        list_genre_2.append(sample[1])
    
    else:
        
        list_genre_1.append(sample[0])   #row in column genres which have only one element will have value alloted to both the lists.
        list_genre_2.append(sample[0])
    
movies["genre_1"]=np.array(list_genre_1)  #creating a column genre_1 and genre_2 from list_genre_1 and list_genre_2 respectively
movies["genre_2"]=np.array(list_genre_2)
print(movies.genres.head(10))    
print(movies.genre_1.head(10))
print(movies.genre_2.head(10))
```

```python
movies_by_segment =movies.groupby(['genre_1','genre_2']) # Write your code for grouping the dataframe here
```

```python
PopGenre = movies_by_segment["gross"].mean().sort_values(ascending=False).head(5).copy()
PopGenre # Write your code for getting the 5 most popular combo of genres here
```

```python
# Write the code for calculating decade here
movies["decade"]=movies["title_year"]-movies["title_year"]%10
```

```python
# Write your code for creating the data frame df_by_decade here 
movies.sort_values("decade",inplace=True)
df_by_decade=movies.groupby("decade")["num_voted_users"].sum()

df_by_decade=pd.DataFrame(df_by_decade)   

df_by_decade.reset_index(inplace=True)
df_by_decade
```

```python
# Write your code for plotting number of voted users vs decade
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(1)
plt.subplot(2,1,1)                                # subplot 1- Bar Graph of num_voted_users by decade
plt.figure(figsize=(15,7))
sns.barplot(x="decade",y="num_voted_users",data=df_by_decade)
plt.title("Bar Graph of num_voted_users by decade")
plt.xlabel("decade")
plt.ylabel("num_voted_users")
plt.close(1)
plt.show()


plt.subplot(2,1,2)                               # subplot 2- Same Bar Graph but on the log value of num_voted_users by decade
plt.figure(figsize=(15,7))                       #            for better clarity
sns.barplot(x="decade",y="num_voted_users",data=df_by_decade)
plt.title("Bar Graph of log(num_voted_users) by decade")
plt.xlabel("decade")                          #Observations- The value of num_voted_users steadily increases from 1910 to 2000 
plt.ylabel("num_voted_users")                                #after 2000, it decreases by substanial amount to the decade 2010
plt.yscale('log')
plt.close(1)                         #2000-2010 decade was marred by economic recession- possible reason for low-budget, 
plt.show()                           #low-grossing movies, also low disposable income and less spending power by the population
                                     #curbing the enthusiasm of movie-goers
                                     #Also- 2000-2010 decade was inspired by the success of Facebook and Twitter, which took away
                                     #majority of the user opinions on their social media websites.
```
